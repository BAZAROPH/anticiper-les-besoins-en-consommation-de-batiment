##### Étape 6: Gestion des outliers
> Nous avons détecté les valeurs aberrantes sur les variables numériques en utilisant la règle de l’IQR (Q1 − 1,5×IQR et Q3 + 1,5×IQR). Les outliers identifiés ont été remplacés par la médiane de la variable correspondante, car celle-ci est robuste aux valeurs extrêmes et permet de conserver toutes les observations. Cette méthode permet ainsi de réduire l’impact des valeurs extrêmes sur les modèles tout en préservant l’ensemble du jeu de données.

```python
for column in numeric_columns:
    #Détection des outliers et remplacement
    outliers_idx = utils.outlier_detect_iqr(building_consumption, column)  # retourner les indices
    building_consumption[column] = building_consumption[column].astype(float)
    median_val = building_consumption[column].median()
    building_consumption.loc[outliers_idx, column] = median_val
```

##### Remplacer les valeurs des colonnes numériques vide ou NaN par la médiane
>Nous avons choisi de remplacer les valeurs manquantes dans les colonnes numériques par la médiane de chaque variable. Cette méthode est robuste aux valeurs extrêmes, permet de conserver toutes les observations du jeu de données et assure que la distribution générale de la variable est peu impactée. Elle est particulièrement adaptée aux variables asymétriques et garantit que les modèles entraînés ne seront pas biaisés par des valeurs manquantes.

```python
for col in building_consumption.columns:
    if building_consumption[col].dtype in ["int64", "float64"]:
        # calculer la médiane seulement si la colonne contient au moins une valeur non-NaN
        if building_consumption[col].notna().any():
            median_value = building_consumption[col].median()
            building_consumption[col] = building_consumption[col].fillna(median_value)
        else:
            print(f"{col} contient uniquement des NaN et ne peut pas être imputée par la médiane")
```

### Feature Engineering

A réaliser : Enrichir le jeu de données actuel avec de nouvelles features issues de celles existantes. 

En règle générale : On utilise la méthode .apply() de Pandas pour créer une nouvelle colonne à partir d'une colonne existante. N'hésitez pas à regarder les exemples dans les chapitres de cours donnés en ressource

```python
# CODE FEATURE ENGINEERING

#Nouvelle Feature 1: Age du batiment
current_year = 2025
building_consumption["BuildingAge"] = building_consumption["YearBuilt"].apply(lambda x: current_year - x)

# Nouvelle feature 2: Nombre de types d’usage par bâtiment
building_consumption["NumPropertyUseTypes"] = building_consumption["ListOfAllPropertyUseTypes"].apply(
    lambda x: len(str(x).split(","))
)

# Nouvelle feature 3: GFA par étage
building_consumption["GFAPerFloor"] = building_consumption["PropertyGFATotal"] / building_consumption["NumberofFloors"]

# Nouvelle feature 4 : consommation d’énergie normalisée par surface
building_consumption["EnergyUsePerArea"] = building_consumption["SiteEnergyUse(kBtu)"] / building_consumption["PropertyGFATotal"]

# Nouvelle feature 5 : émissions selon la surface
building_consumption["EmissionsPerArea"] = building_consumption["TotalGHGEmissions"] / building_consumption["PropertyGFATotal"]

# Nouvelle feature 6 : catégorisation du nombre d’étages
def categorize_floors(x):
    if x <= 3:
        return "LowRise"
    elif x <= 7:
        return "MidRise"
    else:
        return "HighRise"

building_consumption["FloorsCategory"] = building_consumption["NumberofFloors"].apply(categorize_floors)

numeric_columns.extend([
    "BuildingAge",
    "GFAPerFloor",
    "EnergyUsePerArea",
    "EmissionsPerArea",
    "NumPropertyUseTypes"
])

categorical_variables.append("FloorsCategory")
```

###### **Validation croisée avec Random Forest**

```python
X_energy = building_consumption_energy_use.drop(columns=["SiteEnergyUse(kBtu)"])
y_energy = building_consumption_energy_use["SiteEnergyUse(kBtu)"]

rf_model_energy = RandomForestRegressor(
    n_estimators= 100, #plus j'augmente le nombre d'arbre plus le modèle est robuste mais lent
    random_state= 42
)
#choix des métriques
scoring = ["r2", "neg_mean_squared_error", "neg_mean_absolute_error"]

#Validation croisée
cv_energy = cross_validate(
    estimator = rf_model_energy,
    X=X_energy,
    y=y_energy,
    cv=5, #Nombre de fold
    scoring= scoring,
    return_train_score=True
)

# Affichage des résultats
print("=== Validation croisée RandomForest - EnergyUse ===")
cross_metrics = {}
for metric in scoring:
    mean_train = np.mean(cv_energy[f"train_{metric}"])
    std_train = np.std(cv_energy[f"train_{metric}"])
    mean_test = np.mean(cv_energy[f"test_{metric}"])
    std_test = np.std(cv_energy[f"test_{metric}"])
    
    print(f"{metric} → Train : mean={mean_train:.3f}, std={std_train:.3f} | Test : mean={mean_test:.3f}, std={std_test:.3f}")
    cross_metrics[metric] = {
        "mean_train": mean_train,
        "mean_test": mean_test,
    }
metrics_energy["cross_validation"] = cross_metrics
```

###### **Modèle LinearRegression pour energyUse**

```python
# Split en train/test
X_train, X_test, y_train, y_test = train_test_split(X_energy, y_energy, test_size=0.2, random_state=42)

# Initialiser et entraîner le modèle
model_energy = LinearRegression()
model_energy.fit(X_train, y_train)

# Prédiction
y_pred_test = model_energy.predict(X_test)

# Évaluation
mse = mean_squared_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_test)
mae = mean_absolute_error(y_test, y_pred_test)

print("\n===  Modèle LinearRegression SiteEnergyUse ===")
print(f"MSE = {mse:.2f}")
print(f"RMSE = {rmse:.2f}")
print(f"MAE = {mae:.2f}")
print(f"R² = {r2:.2f}")

metrics_energy["LinearRegression"] = {
    "MSE": round(mse, 2),
    "RMSE": round(rmse, 2),
    "MAE": round(mae, 2),
    "R²": round(r2, 2)
}
```

##### Analyse des métriques

Interprétation des métriques en régression
* MSE / RMSE : plus c’est proche de 0, mieux le modèle prédit. → RMSE est plus intuitif car il est dans la même unité que la variable cible.
* MAE : indique l’erreur moyenne absolue. → Plus petit = plus précis. Moins sensible aux valeurs extrêmes que le RMSE.
* R² : mesure la proportion de variance expliquée par le modèle.
    * R² ≈ 1 → modèle très bon.
    * R² ≈ 0 → modèle pas meilleur qu’une simple moyenne.
    * R² < 0 → modèle pire qu’un modèle naïf.

##### Recherche des hyperparams

```python
# Split en train/test
X_train, X_test, y_train, y_test = train_test_split(X_energy, y_energy, test_size=0.2, random_state=42)

# Définition du modèle de base
rf = RandomForestRegressor(random_state=42)

# Grille d’hyperparamètres
param_grid = {
    # Nombre d’arbres dans la forêt
    "n_estimators": [100, 200, 500],
    # Profondeur maximale de chaque arbre
    "max_depth": [None, 10, 20, 30],
    # Plus la valeur est grande, plus l’arbre est "conservateur" (j'évite le surapprentissage)
    "min_samples_split": [2, 5, 10],
    # Nombre minimum d’échantillons requis dans une feuille, c'utile pour limiter la taille des arbres et réduire l’overfitting
    "min_samples_leaf": [1, 2, 4],

    # Nombre de features à considérer pour chaque split
    # "sqrt" = racine carrée du nombre total de features
    # "log2" = logarithme du nombre total de features
    # Permet d’introduire de la diversité entre les arbres
    "max_features": ["sqrt", "log2"]
}

# GridSearchCV
grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    scoring="r2", # Maximiser le le r2
    cv=3,       # 3-fold cross validation
    n_jobs=-1,  # utilise tous les cœurs dispo
    verbose=0
)

# Entraînement
grid.fit(X_train, y_train)

# Meilleurs résultats
print("Meilleurs paramètres :", grid.best_params_)
print("Meilleur score CV :", grid.best_score_)

# Évaluation finale sur test
best_rf = grid.best_estimator_
y_pred_test = best_rf.predict(X_test)

print("\n===  Modèle RandomForestRegressor optimisé (GridSearchCV) ===")
print(f"MSE = {mean_squared_error(y_test, y_pred_test):.2f}")
print(f"R² = {r2_score(y_test, y_pred_test):.2f}")
```

```python
# CODE OPTIMISATION ET INTERPRETATION DU MODELE

X_train, X_test, y_train, y_test = train_test_split(X_energy, y_energy, test_size=0.2, random_state=42)
y_pred = best_rf.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("\n===  Modèle optimisé randomForst siteEnergyUse ===")
print(f"MSE : {mse:.4f}")
print(f"RMSE  : {rmse:.4f}")
print(f"MAE  : {mae:.4f}")
print(f"R²  : {r2:.4f}")
```
